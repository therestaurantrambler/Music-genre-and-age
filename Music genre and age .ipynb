```python
#Create a model using the algorithm decision tree from SkiKit-Learn. The package sklearn has a module called tree. 
#DecisionTreeClassifier is a class that implements the decision tree algorithm.
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

music_df = pd.read_csv('music.csv')
X = music_df.drop(columns=['genre'])
y = music_df['genre']

#Create a new instance of the DecisionTreeClassifier class.This creates the 
#model to be trained.

model = DecisionTreeClassifier()

#Next we need to train the model.Call the fit method on the model. The 
#method takes in two parameters, the input set X and the output set y.

model.fit(X,y)

#Ask the model for a prediction, say What is the music that a 21 male likes. 
#Notice 21 male is not in data set.
#Call the predict method on the model.The method takes in two dimesional array.
#Embed the arrays for predictions [21, 1](21 male) and [20,0](20 female)
#store in a variable called predictions.
#print the value of predictions.

predictions = model.predict([[21,1], [20,0]])
predictions
```

```python
#The algorithm predicted the correct answers
#Need to obtain an accuracy value for the model.
#Split the data into the training data set (70-80% of full data set) and the testing data set (20-30%)
#from the model_selection model import the train_test_split function
#To compare the model predictions with the actual output values in the testing set,
#import the accuracy_score function from the metrics model 

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

#Call the train_test_split function after declaring the variables.
#the function takes in three vaiables, the arrays X, y 
#and a keyword arguement that specifies size of data set (test_size)
#This function returns a tuple.
#Unpack it into 4 variables.Order matters in labeling.


music_df = pd.read_csv('music.csv')
X = music_df.drop(columns=['genre'])
y = music_df['genre']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

#When training the model we want to pass only the training data sets (not the full data set)
#call a new instance of the DecisionTreeClassifier and pass training sets
#Pass the input testing data (X_test) into the predictions
#Compare the predicition values to the y_test data set to determine accuracy
#by calling the accuracy_score function.This function takes in two variables,
#y_test and predictions and returns a value between 0 and 1. Label this value
#score. Display the score.

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
score = accuracy_score(y_test, predictions)
score

#Press ctrl + enter to run the program multiple times. Watch the score change.
```

```python
#Training a model can be very time consuming do not want to have
#to run code multiple times. Save the model to a file for model
#persistence. Allows us to load the trained model for future predictions.
#import joblib object that has methods for saving and loading models

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import joblib

music_df = pd.read_csv('music.csv')
X = music_df.drop(columns = ['genre'])
y = music_df['genre']

model = DecisionTreeClassifier()
model.fit(X,y)

#Call dump on joblib. Give two arguments: the model and the name to store file.

joblib.dump(model,'music-recommender.joblib')

```

```python
# Open the trained music-recommender file by calling load
# pass in the predictions desired 21 male, 20 female
#print predictions

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import joblib

joblib.load('music-recommender.joblib')
predictions = model.predict([ [21,1], [20,0]])
predictions
```

```python
#Visualize the model. Import the tree object that has a method for
# for visualizing tree as a graph.

import pandas as pd
from sklearn.tree import DecisionTreeClassifier

from sklearn import tree

music_df = pd.read_csv('music.csv')
X = music_df.drop(columns = ['genre'])
y = music_df['genre']

model = DecisionTreeClassifier()
model.fit(X,y)

# After model is trained, call the export_graphviz method on the tree object.
# The method takes in arguements: the model and name of outputfile
#with feature names--the array of inputs
#class names-- array of UNIQUE output lables sorted ALPHABETICALLY
#For rounded corners set = True
# filled = True fill nodes with colors
#label = all, every node has labels that can be read

tree.export_graphviz(model, out_file = 'music-recommender.dot',
                    feature_names = ['age','gender'],
                    class_names = y.unique(),
                    label = 'all',
                    rounded = True,
                    filled = True)

#The exported file is in the folder. Open the .dot file in a VS Code window.
#The file is a data text file. To visualize, we need to install and extention
#in VS Code.

#To install, the extentions click the Extentions icon on left in VS Code.
#Search for dot. Select the Graph Viz (dot) language Support for VS Code 
#by Joao Pinto (Stephanvs did not have preview on side available).
#Install extention and reload VS Code. Reopen VS Code. Click on three
#dots in upper right. Selct "Open Preview to Side".
```
